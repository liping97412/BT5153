# Scraping data from quora

```python
import urllib.request
url = "https://www.quora.com/topic/Finance"
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
#get the html from the provided url
html = response.read().decode('unicode_escape')
```

```python
import bs4
from bs4 import BeautifulSoup
```

```python
#call Beautifulsoup to convert plain html to a structured format
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.prettify())
```

```python
def extract_question_text(soup): 
    questions = []
    for div in soup.find_all(name="div", attrs={"class":"story_title_container"}):
        question = div.find_all(name="span", attrs={"class":"ui_qtext_rendered_qtext"})       
        for q in question:
            questions.append(q.text.strip())
    return(questions)
questions = extract_question_text(soup)
```
